### Pandas
- csv파일 불러오기
- 총계 구하기 : sum()
  - axis = 0 : 열별 합계
  - axis = 1 : 행별 합계
- 평균 구하기 : mean()
  - axis = 0 : 열별 합계
  - axis = 1 : 행별 합계
- loc : 행 또는 열의 이름으로 데이터 추출
- iloc : 행 또는 열의 인덱스로 데이터 추출
- drop : 데이터 삭제
- fillna(채울 값, inplace = ?) : 결측값 채우기
  - inplace = True로 지정하면 바로 적용됨
- min
  - axis = 0 : 열별 합계
  - axis = 1 : 행별 합계
- max
  - axis = 0 : 열별 합계
  - axis = 1 : 행별 합계

```python
import pandas as pd

#csv파일 불러오기
data = pd.read_csv("C:/BigData_MachineLearning/dataset/score.csv", index_col = "과목" ,encoding = "euc-kr")
data
```

```python
# 학급별 총계 구해보기
# sum()
# axis = 0 : 열별 합계
# axis = 1 : 행별 합계
print(data.sum())
print(data["1반"].sum())
```

```python
print(data.sum(axis=1))
```

```python
# 과목별 총합 컬럼 추가
data["총합"] = data.sum(axis = 1)
data
```

```python
# 과목별 평균 컬럼 추가
data["평균"] = data[["1반","2반","3반","4반"]].mean(axis=1)
data
```

```python
# 컬럼 삭제
data = data.drop(columns="평균")
data
```

```python
# 과목별 평균 컬럼 추가
# loc 사용
data["평균"] = data.loc["수학":"과학","1반":"4반"].mean(axis=1)
data
```

```python
data.loc["반평균"] = data.iloc[:,:4].mean()
data
```

```python
# 결측값 0으로 채우기
data.fillna(0, inplace=True)
data
```

```python
# 가장 큰 값 가져오기 : max()
# 가장 작은 값 가져오기 : min()
# axis = 0 : 열별
# axis = 1 : 행별
print(data.max())
print(data.max(axis = 1))
print(data.min())
print(data.min(axis = 1))
```
