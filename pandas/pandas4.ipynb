```python
import pandas as pd
```

```python
crime_2015 = pd.read_csv("C:/BigData_MachineLearning/dataset/2015.csv", index_col="관서명", encoding="euc-kr")
crime_2016 = pd.read_csv("C:/BigData_MachineLearning/dataset/2016.csv", index_col="관서명", encoding="euc-kr")
crime_2017 = pd.read_csv("C:/BigData_MachineLearning/dataset/2017.csv", index_col="관서명", encoding="euc-kr")
crime_2015
```

```python
result_set = crime_2015.loc[crime_2015["구분"]=="발생건수"]
result_set
```

```python
result_set["2015총계"] = crime_2015.loc[crime_2015["구분"]=="발생건수"].sum(axis=1)
result_set
```

```python
result_set.drop(columns=["구분","살인","강도","강간·강제추행","절도","폭력"],inplace=True)
result_set
```

```python
result_set["2016총계"] = crime_2016.loc[crime_2016["구분"]=="발생건수"].sum(axis=1)
result_set["2017총계"] = crime_2017.loc[crime_2017["구분"]=="발생건수"].sum(axis=1)
result_set
```

```python
result_set["2015-2016 증감률"] = (result_set["2016총계"]-result_set["2015총계"])/result_set["2015총계"]*100
result_set["2016-2017 증감률"] = (result_set["2017총계"]-result_set["2016총계"])/result_set["2016총계"]*100
result_set
```
